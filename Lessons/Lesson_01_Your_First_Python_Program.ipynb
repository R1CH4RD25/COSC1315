## üìå Key Concepts Recap
- `print(...)` outputs text to the screen.
- Python is **case-sensitive**: `print` is correct, `Print` is not.
- Strings go in matching quotes: `'text'` or `"text"`.
- Use escaping or alternate quotes when your string contains quotes.
